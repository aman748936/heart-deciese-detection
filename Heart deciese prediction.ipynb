# Train test split 

```python
from sklearn.model_selection import train_test_split

predictors = dataset[["Age","RestBP","Chol","MaxHR","ExAng","ChestPain_asymptomatic","RestECG"]]
target = dataset["MaxHR"]

X_train,X_test,Y_train,Y_test = train_test_split(predictors,target,test_size=0.20,random_state=0)
```

# Satrting the fitting the Model

```python
from sklearn.metrics import accuracy_score
dataset.fillna(0)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
 model = lr.fit(X_train,Y_train)
y_pred_lr = lr.predict(X_test)
```

```python
score_lr = accuracy_score(y_pred_lr,Y_test)
```

```python
score_lr
```

# using naive  bayes algorithm


```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()

nb.fit(X_train,Y_train)

Y_pred_nb = nb.predict(X_test)
```

```python
Y_pred_nb.shape
```

```python
score_nb = accuracy_score(Y_pred_nb,Y_test)
score_nb
```
